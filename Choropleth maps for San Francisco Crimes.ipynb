```python
df2 = df['PdDistrict'].value_counts()
df2
```

```python
df2 = pd.DataFrame(df2)
df2
```

```python
#df2 = df2.reset_index()
#df2['Neighborhood'] = df2['index']
#df2['Count'] = df2['PdDistrict']
df2 = df2.drop('index',axis=1)
df2 = df2.drop('level_0',axis=1) 
df2
```

```python
world_geo = r'san-francisco.json' # geojson file

# create a plain world map
world_map = folium.Map(location=[0, 0], zoom_start=12, tiles='Mapbox Bright')
```

```python
world_geo = r'F://san-francisco.geojson'


threshold_scale = np.linspace(df2['Count'].min(),
                              df2['Count'].max(),
                              6, dtype=int)
threshold_scale = threshold_scale.tolist() # change the numpy array to a list
threshold_scale[-1] = threshold_scale[-1] + 1 # make sure that the last value of the list is greater than the maximum

# let Folium determine the scale.
world_map = folium.Map(location=[0, 0], zoom_start=12, tiles='Mapbox Bright')
world_map.choropleth(
    geo_data=world_geo,
    data=df2,
    columns=['Neighborhood', 'Count'],
    key_on='feature.properties.name',
    threshold_scale=threshold_scale,
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crimes in San Francisco',
    reset=True
)
world_map
```
